```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
credit = pd.read_csv('default of credit card clients.csv', header = 1)
```

```python
credit.describe()
```

```python
credit.info()
```

```python
credit.isnull()
```

```python
creditDF = credit
```

```python
# Cleaning up of the DF

creditDF.columns = [x.lower() for x in creditDF.columns]
creditDF = creditDF.rename(index = str, columns = {"pay_0" : "pay_1"})
creditDF = creditDF.rename(index = str, columns = {"default payment next month" : "defaults"})
creditDF = creditDF.drop( 'id', axis =1)
creditDF.columns
```

```python
# Grouping age into bins using the cut()

creditDF['age_bin'] = pd.cut(creditDF['age'], 6, labels=["20", "30", "40", "50", "60", "70"])
creditDF['age_bin']
```

```python
# Grouping limit_bal into bins using the cut()

creditDF['limit_bal_bin'] = pd.cut(creditDF['limit_bal'],10, 
                                   labels=["10K","20K", "30K", "40K", "50K", "60K", "70K","80K","90K","100K"])

creditDF['limit_bal_bin']
```

```python
# first 10 rows of the DF
creditDF[:10]
```

```python
# Pivot tables
creditDF.pivot_table('defaults', 'limit_bal_bin',aggfunc = len, margins = True)
```

```python
creditDF.pivot_table('defaults', 'education',aggfunc = len, margins = True)
```

```python
creditDF.pivot_table('defaults', 'sex','marriage',aggfunc = len, margins = True)
```

```python
creditDF.pivot_table('defaults', 'age_bin',aggfunc = len, margins = True)
```

```python
sns.catplot('limit_bal_bin', data = creditDF, kind = 'count')
```

```python
# Visualization using plt.subplots - Defaults vs others
fig, ax = plt.subplots(1,3)
fig.set_size_inches(20,5)

d = creditDF.groupby(['defaults', 'marriage']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[0])

d = creditDF.groupby(['defaults', 'sex']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[1])

d = creditDF.groupby(['defaults', 'age_bin']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[2])
```

```python
# Visualization using plt.subplots - Defaults vs others
fig, ax = plt.subplots(1,2)
fig.set_size_inches(20,5)

d = creditDF.groupby(['defaults', 'education']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[0])

d = creditDF.groupby(['defaults', 'limit_bal_bin']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[1])
```

```python
# Visualization using plt.subplots - limit_bin_bal vs age_bin & education
fig, ax = plt.subplots(1,2)
fig.set_size_inches(20,5)

d = creditDF.groupby(['limit_bal_bin', 'education']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[0])

d = creditDF.groupby(['limit_bal_bin', 'age_bin']).size()
p = d.unstack(level=1).plot(kind='bar', ax=ax[1])
```

```python
# Correlation matrix
corrMat = creditDF[['limit_bal','bill_amt1', 'bill_amt2','bill_amt3','bill_amt4','bill_amt5','bill_amt6',
                  'pay_amt1','pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']].corr()

corrMat
```

```python
# Heatmap for the Correlation matrix 

sns.heatmap(corrMat, xticklabels=corrMat.columns, yticklabels=corrMat.columns, annot=True)
```

```python
# Covariance Matrix

covMat = creditDF[['limit_bal','bill_amt1', 'bill_amt2','bill_amt3','bill_amt4','bill_amt5','bill_amt6',
                  'pay_amt1','pay_amt2', 'pay_amt3', 'pay_amt4', 'pay_amt5', 'pay_amt6']].cov()
covMat
```
